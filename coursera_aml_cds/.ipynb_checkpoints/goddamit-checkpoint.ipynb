```python
transactions['item_units'] = transactions.groupby(['item_id'])['item_cnt_day'].transform(np.sum)
transactions['item_block_units'] = transactions.groupby(['item_id','date_block_num'])['item_cnt_day'].transform(np.sum)
transactions['item_mean_units_block'] = transactions.groupby(['item_id'])['item_block_units'].transform(np.mean)
transactions['item_day_units'] = transactions.groupby(['item_id','date'])['item_cnt_day'].transform(np.sum)
transactions['item_mean_units_day'] = transactions.groupby(['item_id'])['item_day_units'].transform(np.mean)
transactions['item_max_units_block'] = transactions.groupby(['item_id'])['item_block_units'].transform(np.max)
transactions['item_min_units_block'] = transactions.groupby(['item_id'])['item_block_units'].transform(np.min)
transactions['item_max_units_day'] = transactions.groupby(['item_id'])['item_day_units'].transform(np.max)
transactions['item_min_units_day'] = transactions.groupby(['item_id'])['item_day_units'].transform(np.min)
```

```python
transactions['item_turnover'] = transactions.groupby(['item_id'])['turnover'].transform(np.sum)
transactions['item_block_turnover'] = transactions.groupby(['item_id','date_block_num'])['turnover'].transform(np.sum)
transactions['item_mean_turnover_block'] = transactions.groupby(['item_id'])['item_block_turnover'].transform(np.mean)
transactions['item_day_turnover'] = transactions.groupby(['item_id','date'])['turnover'].transform(np.sum)
transactions['item_mean_turnover_day'] = transactions.groupby(['item_id'])['turnover'].transform(np.mean)
transactions['item_max_turnover_block'] = transactions.groupby(['item_id'])['item_block_turnover'].transform(np.max)
transactions['item_min_turnover_block'] = transactions.groupby(['item_id'])['item_block_turnover'].transform(np.min)
transactions['item_max_turnover_day'] = transactions.groupby(['item_id'])['item_day_turnover'].transform(np.max)
transactions['item_min_turnover_day'] = transactions.groupby(['item_id'])['item_day_turnover'].transform(np.min)
```

```python
transactions['item_days_of_activity'] = transactions.groupby(['item_id'])['date'].transform("nunique")
transactions['item_blocks_of_activity'] = transactions.groupby(['item_id'])['date_block_num'].transform("nunique")

def get_number_of_days_since_start(day,month, year):
    days = 0
    if year == 2015:
        days = 365
    def is_even(num):
        return num % 2 == 0
    half_of_month = int(month/2)
    even = (30*half_of_month) + (31*half_of_month)
    if is_even(month):
        days = days + even - 30 - day
    else:
        days = days + even + day
    return days

transactions['item_days_since_start'] = transactions.apply(lambda row: get_number_of_days_since_start(row['day'],row['month'], row['year']),axis=1)

def get_average_days_between_sales(days):
    days = sorted(np.unique(days))
    if len(days) == 0:
        return 9999
    if len(days) == 1:
        return 999
    return np.mean(np.ediff1d(days)) / len(days)

average_days_between_sales = transactions.groupby(['item_id'])['item_days_since_start'].apply(list).apply(lambda x: get_average_days_between_sales(x))

transactions['item_mean_day_between_activity'] = transactions['item_id'].map(average_days_between_sales)


def get_max_stretch_without_sales_days(days):
    days = np.unique(days)
    max_stretch = 0
    len_days = len(days)
    for index,day in enumerate(sorted(days)):
        if index == len_days - 1:
            return max_stretch
        next_day = days[index+1]
        stretch = next_day - day
        if stretch > max_stretch:
            max_stretch = stretch
            

        
max_stretch_without_sales_day = transactions.groupby(['item_id'])['item_days_since_start'].apply(list).apply(lambda x: get_max_stretch_without_sales_days(x))

transactions['item_longest_stretch_days_without_activity'] = transactions['item_id'].map(max_stretch_without_sales_day)
```

```python
gc.collect()

def get_max_stretch_without_sales_block(blocks):
    blocks = np.unique(blocks)
    max_stretch = 0
    len_blocks = len(blocks)
    for index,block in enumerate(sorted(blocks)):
        if index == len_blocks - 1:
            return max_stretch
        next_block = blocks[index+1]
        stretch = next_block - block
        if stretch > max_stretch:
            max_stretch = stretch
            

        
item_longest_stretch_blocks_without_activity = transactions.groupby(['item_id'])['date_block_num']\
                                    .apply(list).apply(lambda x: get_max_stretch_without_sales_block(x))

transactions['item_longest_stretch_blocks_without_activity'] = transactions['item_id'].map(item_longest_stretch_blocks_without_activity)



def get_longest_stretch(following_pairs, n=1,new_n=1):
    #print("following_pairs", following_pairs, " n: ", n, " new_n: ", new_n)
    len_pairs = len(following_pairs)
    if len_pairs == 0:
        return 0
    if len_pairs == 1:
        if new_n > n:
            return new_n
        return n
    if following_pairs[1][0] == following_pairs[0][1]:
        new_n+=1
    else:
        if new_n > n:
            n=new_n
        new_n=1
    return get_longest_stretch(following_pairs[1:], n,new_n)


assert(get_longest_stretch([]) == 0)
assert(get_longest_stretch([[1, 2], [2,3], [3, 4], [4,5] ,[8,9], [11, 12]]) == 4)
assert(get_longest_stretch([[-1, 0],[1, 2], [2,3], [3, 4], [4,5] ,[8,9], [11, 12]]) == 4)
assert(get_longest_stretch([[1, 2], [4,5] ,[8,9], [9,10],[10, 11]]) == 3)
assert(get_longest_stretch([[1, 2], [4,5] ,[8,9], [9,10],[10, 11],[20, 21], [25,26]]) == 3)
assert(get_longest_stretch([[1, 2], [4,5] ,[8,9], [9,10],[10, 11],[14, 15], [15,16]]) == 3)
assert(get_longest_stretch([[1, 2], [4,5] ,[8,9], [9,10],[10, 11],[14, 15], [15,16],[18,19] ,[22,23], [23,24],[24, 25]]) == 3)
assert(get_longest_stretch([[1, 2], [4,5], [7, 8]]) == 1)
assert(get_longest_stretch([[1, 2], [5, 6], [6, 7], [7,8], [14, 15]]) == 3)
assert(get_longest_stretch([[1, 2], [5, 6], [6, 7], [7,8], [14, 15], [15, 16]]) == 3)
assert(get_longest_stretch([[1, 2], [5, 6], [6, 7], [7,8], [14, 15], [15, 16], [16, 17]]) == 3)
assert(get_longest_stretch([[1, 2], [5, 6], [6, 7], [14, 15], [15, 16], [16, 17]]) == 3)
assert(get_longest_stretch([[1, 2], [5, 6], [6, 7], [7,8], [14, 15], [20, 21], [21, 22], [22,23],[23,24]]) == 4)


def get_following_pairs(pairs):
    pairs = np.unique(pairs)
    len_pairs = len(pairs)
    following = []
    for index,pair in enumerate(sorted(pairs)):
        if index == len_pairs - 1:
            return following
        next_pair = pairs[index+1]
        if next_pair == pair + 1:
            following.append([pair, next_pair])
        
assert(get_following_pairs([1,2,5,6,7,8,9,11,12,15]) == [[1, 2], [5, 6], [6, 7], [7, 8], [8, 9], [11, 12]])
assert(get_following_pairs([1,2,5,6,7,10]) == [[1, 2], [5, 6], [6, 7]])
assert(get_following_pairs([1,2,4,5,7,9,10]) == [[1, 2], [4, 5], [9,10]])
assert(get_following_pairs([1,2,4,5,7,9,10,11,12,15]) == [[1, 2], [4, 5], [9,10],[10,11],[11,12]])


item_longest_stretch_block_with_activity = transactions.groupby(['item_id'])['date_block_num']\
                                    .apply(list).apply(lambda x: get_longest_stretch(get_following_pairs(x)))

transactions['item_longest_stretch_block_with_activity'] = transactions['item_id'].map(item_longest_stretch_block_with_activity)


item_number_of_consecutive_days_with_activity = transactions.groupby(['item_id'])['item_days_since_start']\
                                    .apply(list).apply(lambda x: len(get_following_pairs(x)))
    
transactions['item_number_of_consecutive_days_with_activity'] = transactions['item_id'].map(item_number_of_consecutive_days_with_activity)
```

```python
def get_units_between_first_and_last(units):
    return np.max(units) - np.min(units)

item_days_between_start_and_first_activity = transactions.groupby(['item_id'])['item_days_since_start'].apply(list).apply(lambda x: get_days_between_first_and_last(x))
transactions['item_days_between_start_and_first_activity'] = transactions['item_id'].map(item_days_between_start_and_first_activity)

item_blocks_between_start_and_first_activity = transactions.groupby(['item_id'])['date_block_num'].apply(list).apply(lambda x: get_days_between_first_and_last(x))
transactions['item_blocks_between_start_and_first_activity'] = transactions['item_id'].map(item_blocks_between_start_and_first_activity)
```

```python
transactions['item_first_block'] = transactions.groupby('item_id')['date_block_num'].transform(np.min)
transactions['item_last_block'] = transactions.groupby('item_id')['date_block_num'].transform(np.max)
transactions['item_first_day'] = transactions.groupby('item_id')['item_days_since_start'].transform(np.min)
transactions['item_last_day'] = transactions.groupby('item_id')['item_days_since_start'].transform(np.max)

item_activity_on_all_blocks = transactions.groupby('item_id')['date_block_num'].nunique().apply(lambda x: x==number_of_blocks)
transactions['item_activity_on_all_blocks'] = transactions['item_id'].map(item_activity_on_all_blocks)
```

```python
transactions['item_mean_price'] = transactions.groupby('item_id')['item_price'].transform(np.mean)
transactions['item_mean_price_block'] = transactions.groupby(['item_id', 'date_block_num'])['item_price'].transform(np.mean)
transactions['item_min_price'] = transactions.groupby('item_id')['item_price'].transform(np.min)
transactions['item_max_price'] = transactions.groupby('item_id')['item_price'].transform(np.max)
transactions['item_number_different_prices'] = transactions.groupby('item_id')['item_price'].transform('nunique')
transactions['item_price_amplitude'] = ((transactions['item_max_price'] - transactions['item_min_price'] ) / transactions['item_min_price']) * 100
transactions['category_mean_price'] = transactions.groupby('item_category_id')['item_price'].transform(np.mean)
transactions['item_deviation_mean_category_price'] =  ((transactions['item_mean_price'] - transactions['category_mean_price'] ) / transactions['category_mean_price']) * 100
```

```python
transactions['is_first_two_blocks'] = transactions['date_block_num'].isin([transactions['item_first_block']+1,transactions['item_first_block']+2])
transactions['is_last_two_blocks'] = transactions['date_block_num'].isin([transactions['item_last_block']-1,transactions['item_last_block']])


item_first_two_blocks_units = transactions.groupby(['item_id','is_first_two_blocks'], as_index=False)['item_cnt_day'].sum()
item_first_two_blocks_units = item_first_two_blocks_units[item_first_two_blocks_units['is_first_two_blocks'] == True].set_index('item_id').iloc[:,1]
transactions['item_first_two_blocks_units'] = transactions['item_id'].map(item_first_two_blocks_units)

item_last_two_blocks_units = transactions.groupby(['item_id','is_last_two_blocks'], as_index=False)['item_cnt_day'].sum()
item_last_two_blocks_units = item_last_two_blocks_units[item_last_two_blocks_units['is_last_two_blocks'] == True].set_index('item_id').iloc[:,1]
transactions['item_last_two_blocks_units'] = transactions['item_id'].map(item_last_two_blocks_units)

transactions['item_fluctuation_units_first_last_blocks'] =  ((transactions['item_first_two_blocks_units'] - transactions['item_last_two_blocks_units'] ) / \
                                                             transactions['item_first_two_blocks_units']) * 100 * -1


item_first_two_blocks_mean_price = transactions.groupby(['item_id','is_first_two_blocks'], as_index=False)['item_price'].mean()
item_first_two_blocks_mean_price = item_first_two_blocks_mean_price[item_first_two_blocks_mean_price['is_first_two_blocks'] == True].set_index('item_id').iloc[:,1]
transactions['item_first_two_blocks_mean_price'] = transactions['item_id'].map(item_first_two_blocks_mean_price)

item_last_two_blocks_mean_price = transactions.groupby(['item_id','is_last_two_blocks'], as_index=False)['item_price'].mean()
item_last_two_blocks_mean_price = item_last_two_blocks_mean_price[item_last_two_blocks_mean_price['is_last_two_blocks'] == True].set_index('item_id').iloc[:,1]
transactions['item_last_two_blocks_mean_price'] = transactions['item_id'].map(item_last_two_blocks_mean_price)

transactions['item_fluctuation_price_first_last_blocks'] =  ((transactions['item_first_two_blocks_mean_price'] - transactions['item_last_two_blocks_mean_price'] ) / \
                                                             transactions['item_first_two_blocks_mean_price'])  * 100 * -1
```

```python
transactions['item_share_of_total_units'] = transactions['item_units'] * 100 / total_sales 

transactions['share_of_total_turnover'] = transactions['item_turnover'] * 100 / total_turnover

transactions['category_units'] = transactions.groupby('item_category_id')['item_cnt_day'].transform(np.sum)
transactions['item_share_of_category_units'] = transactions['item_units'] * 100 / transactions['category_units']

transactions['category_turnover'] = transactions.groupby('item_category_id')['turnover'].transform(np.sum)
transactions['item_share_of_category_turnover'] = transactions['item_turnover'] * 100 / transactions['category_turnover']
```
